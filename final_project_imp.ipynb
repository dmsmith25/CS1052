```python
pickle_in = open('2011-2021-data-final', 'rb')
dataset = pickle.load(pickle_in)

dataset_sub = dataset

df_features = dataset_sub.iloc[:, 0:-6]
df_features["Tournament_Game"] = dataset_sub["Tournament_Game"]
df_labels = dataset_sub.iloc[:, -6:-1]

df_mean = np.mean(df_features, axis=0)
df_std = np.std(df_features, axis=0)

df_features = (df_features - np.mean(df_features, axis=0)) / (np.std(df_features, axis=0))

df_features = df_features.drop(columns=["Tournament_Game", "Away_Games Played", "Home_Games Played"], axis=1)

df_features
```

```python
table = LSH.generateTable(dataset, df_features, seeds, seeds_t, df_labels, t, r)
```

```python
pickle_in = open('CBB_Data_2022_full_final', 'rb')
testing_data = pickle.load(pickle_in)

testing_data = testing_data.drop(columns=["Home_SOS", "Away_SOS"], axis=1)

testing_data = testing_data.apply(pd.to_numeric, errors='coerce')

testing_data = testing_data.reset_index(drop=True)

testing_features = testing_data.iloc[:, :-2]
testing_labels = testing_data.iloc[:, -2:]

testing_features = testing_features.reset_index(drop=True)
testing_labels = testing_labels.reset_index(drop=True)

testing_features = (testing_features - np.mean(testing_features,axis=0)) / (np.std(testing_features, axis=0))

testing_features
```

```python
preds = []
res = []
score = []


for c in range(3000):
    entries = []

    for q in range(t):
        test_hash = []
        for n in range(r):
            test_hash.append(LSH.cosine_hash_with_seed(testing_features.iloc[c, :], seeds[q], r))

        entries.append(tuple(test_hash))


    home_scores = []
    away_scores = []


    for a in range(t):
        val = LSH.hash_tuple_with_seed(entries[a], len(dataset) - 1, seeds_t[a])
        data = table[a][val]
        if len(data) > 0:
            for stat in data:
                home_scores.append(int(stat['Home_Result']))
                away_scores.append(int(stat["Away_Result"]))


    if len(home_scores) > 0:
        home_mean = np.median(home_scores)
        away_mean = np.median(away_scores)

        pred_res = home_mean - away_mean
        real_res = testing_labels.iloc[c, 1]
        spread = testing_labels.iloc[c, 0]

        preds.append(pred_res)
        res.append(real_res)

        if (real_res + spread) * (pred_res + spread) > 0:
            score.append(1)
        else:
            score.append(0)
```

```python
mse = np.mean((np.array(preds) - np.array(res)) ** 2)
acc = np.sum(score) / len(score)

print("MSE: " + str(mse))
print("ACC: " + str(acc))
```
