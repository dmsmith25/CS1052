```python
def checkTGames(x):
    if int(x["Date"][6]) == 4 or int(x["Tournament_Game"]) == 1:
        return 1
    else:
        return 0
    
data_raw["Tournament_Game"] = data_raw.apply(checkTGames, axis=1)


data_raw = data_raw.dropna(axis=0)

data_raw
```
